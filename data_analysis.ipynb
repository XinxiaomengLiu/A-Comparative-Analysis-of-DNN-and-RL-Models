```python
counts = data.groupby('id').size().tolist()

# print the list of counts
print(counts)
```

```python
counts_stats = {}
for count in counts:
    if count in counts_stats:
        counts_stats[count] += 1
    else:
        counts_stats[count] = 1
print(counts_stats)
```
